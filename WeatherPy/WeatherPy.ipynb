```python
import matplotlib.pyplot as plt
import requests
from scipy import stats
from scipy.stats import linregress
import pandas as pd
from api_keys import weather_api_key
import json
import numpy as np
from citipy import citipy
import time

#output file
output_data_csv = "Output/weather_data.csv"
```

```python
#make lists
cities = []
coordinates = []

#select range 
latitude_range= (-90, 90)
longitude_range = (-180, 180)

#find random values for the coordinates within the range
latitudes= np.random.uniform(low=-90, high = 90, size= 1200)
longitudes= np.random.uniform(low=-180, high = 180, size= 1200)
coordinates= list(zip(latitudes, longitudes))

for coordinate in coordinates:
    city= citipy.nearest_city(coordinate [0], coordinate [1]).city_name
    if city not in cities:
        cities.append(city)

list_data= list(zip(cities,coordinates))
list_data_2 = pd.DataFrame(list_data, columns=["City", "Coordinates"])
list_data_2.head()
```

```python
# set lists for the dataframe
city_list = []
lat= []
lng=[]
max_temp=[]
humidity=[]
cloudiness=[]
wind_speed=[]
country=[]
date=[]
```

```python
# loops for creating dataframe columns
for index, row in list_data_2.iterrows():
    time.sleep(1)
    city=row["City"]
    url= f"http://api.openweathermap.org/data/2.5/weather?q={city}&units=imperial&appid={weather_api_key}"
    response = requests.get(url).json()
    print(json.dumps(response,indent=4))
    

    if response['cod']==200:
        lat.append(response['coord']['lat'])
        lng.append(response ['coord']['lon'])
        max_temp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        city_list.append(city)

    else:
        print("City not found. Skipping...")
```

```python
weather_data = ({"City":city_list,
    "Cloudiness":cloudiness,
    "Country":country,
    "Date":date,
    "Humidity":humidity,
    "Lat":lat,
    "Lng":lng,
    "Max Temp":max_temp,
    "Wind Speed":wind_speed})
```

```python
weather_df = pd.DataFrame(weather_data)

pd.DataFrame.to_csv(weather_df, 'weather_df.csv')
weather_df.head()
```

```python
plt.scatter(weather_df["Lat"],weather_df["Max Temp"])
plt.title("City Latitude vs. Max Temperature")
plt.xlabel("Latitude")
plt.ylabel("Temperature in (F)")
plt.savefig("LatTemp.png")
plt.show()
print(f'As noted through this graph, we can see that through this sample there were more Northern cities. It almost creates a normal distribution curve as we see on both sides it starts at lower temperatures and as it gets to the middle of the equator, the temperature rises.')
```

```python
# Create Northern and Southern Hemisphere DataFrames
northern = weather_df.loc[weather_df["Lat"] >= 0]
southern = weather_df.loc[weather_df["Lat"] < 0]
```

```python
x_values = northern["Lat"]
y_values = northern["Max Temp"]
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)

#regression values
regression_values = x_values * slope + intercept

#Line equation
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(line_eq)

#plotting and formatting

plt.scatter(x_values, y_values)
plt.plot(x_values,regression_values,"r-")

plt.title(f'Northern Hemisphere Latitude vs Temp')
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
print(f"The r-squared is: {rvalue**2}")
plt.show()
plt.savefig("NorthLatTemp.png")
print(f'As noted through this graph, we can see that the Northern Hemisphere tends to fall in temperature the closer we get to the equator. The r-squared is under .7 which means the temperature is not closely correlated with the temperature. This sample does show more correlation to the lattiude than the sample with the Southern Hemisphere')
```
